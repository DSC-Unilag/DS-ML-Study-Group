```python
df_diabetes = df_diabetes.drop(['sex','s4'], axis=1) #drop irrevelant features in the DataFrame
```

# CREATING DEPENDENT AND INDEPENDENT VARIABLE

```python
X = df_diabetes.drop(['target'], axis=1) #The dependent Variable 
y = df_diabetes['target'].values #Independent Variable
```

```python
X
```

```python
y
```

# SPLITTING BOTH DEPENDENT AND INDEPENDENT TO TRAIN AND TEST SETS

```python
from sklearn.model_selection import train_test_split

X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

# SCALING

All features(dependent variables) were been standardized(scaled) in the dataset apart from the target, We can choose to scale the target(independent variable) or not if only we're using Linear Regression because Linear Regression will automatically scale the target(independent variable). Other models like KNN, Random Forest, Decision Trees does the same thing. if we use SVM without scaling the target variable the model will perform badly 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
y_train = scaler.fit_transform(y_train.reshape(-1,1))
y_test = scaler.transform(y_test.reshape(-1,1))
```

# MODEL TRAINING

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred = scaler.inverse_transform(model.predict(X_test)) #reverts the scaling
```
