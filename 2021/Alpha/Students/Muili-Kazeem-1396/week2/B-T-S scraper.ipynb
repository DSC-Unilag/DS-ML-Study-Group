```python
import requests #import requests module
from bs4 import BeautifulSoup as bs #import beautifulsoup
import pandas as pd
import os
```

```python
#load booktoscrape website
bts_website = requests.get("http://books.toscrape.com/")

#convert the webpage to a beautifulsoup object using the code below
bts = bs(bts_website.content)
```

```python
def each_page(content):
    all_list = content.select("li.col-xs-6.col-sm-4.col-md-3.col-lg-3")
    this_page = []
    start = "http://books.toscrape.com/"
    for each_list in all_list:
        name = [each_list.h3.a["title"]]
        checkURL = each_list.div.a.img["src"]
        if checkURL.startswith('../'):
            imageURL = [f'{start}{checkURL.lstrip("../")}']
        else:
            imageURL = [f'{start}{checkURL}']
        ini_price = each_list.select("div.product_price > p.price_color")
        for price in ini_price:
            price = [price.get_text()]
        ini_rating = each_list.p["class"]
        rating = [f"{ini_rating[0]} {ini_rating[1]}"]
        this_page = this_page + [name + imageURL + price + rating]
    return this_page
        
page = each_page(bts)

for i in range(2, 51):
    to_next_page = f"http://books.toscrape.com/catalogue/page-{i}.html"
    bts_website = requests.get(to_next_page)
    bts = bs(bts_website.content)
    page = page + each_page(bts)

headers = ['title', 'imageURL', 'price', 'rating']
df = pd.DataFrame(page, columns=headers)
df.head(30)
```
