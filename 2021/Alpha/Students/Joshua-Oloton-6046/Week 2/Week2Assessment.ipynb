```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
```

```python
index_page = requests.get("http://books.toscrape.com/catalogue/category/books_1/index.html")
webpage = bs(index_page.content)
```

```python
# names
```

### Get book info

```python
names = webpage.select("div ol li article h3 a")
names =[names[i]["title"] for i in range(len(names))]
```

```python
image_url = webpage.select("div ol li article img")
image_url = [image_url[i]["src"] for i in range(len(image_url))]
```

```python
price = webpage.select("p.price_color")
price = [price[i].text for i in range(len(price))]
```

```python
rating = webpage.select("article.product_pod p.star-rating")
rating = [rating[i]["class"][1]+" Stars" for i in range(len(rating))]
```

```python
names1, image_url_1, price1, rating1 = [], [], [], []

base_url = "http://books.toscrape.com/catalogue/category/books_1/"
for page in range(2,51):
    # Get URL for all pages
    next_page = base_url+f"page-{page}.html"
    next_page = requests.get(next_page)
    next_webpage = bs(next_page.content)
    # Get book names for the remaining pages
    names1 = [next_webpage.select("div ol li article h3 a")[i]["title"] for i in range(len(next_webpage.select("div ol li article h3 a")))]
    # All name info for each page is stored in names1 which is then appended to names list before the next loop(page)
    names+=names1
    names1 = []
    # Get image URLs the remaining pages
    image_url_1 = [next_webpage.select("div ol li article img")[i]["src"] for i in range(len(next_webpage.select("div ol li article img")))]
    image_url+=image_url_1
    image_url_1 = []
    # Get price for the remaining pages
    price1 = [next_webpage.select("p.price_color")[i].text for i in range(len(next_webpage.select("p.price_color")))]
    price+=price1
    price1 = []
    # Get rating for the remaining pages
    rating1 = [next_webpage.select("article.product_pod p.star-rating")[i]["class"][1]+" Stars" for i in range(len(next_webpage.select("article.product_pod p.star-rating")))]
    rating+=rating1
    rating1 = []
```

```python
print(len(names))
print(len(image_url))
print(len(price))
print(len(rating))
```

```python
df = pd.DataFrame({'Names':names, 'Image_Url':image_url, 'Price':price, 'Rating':rating })
```
