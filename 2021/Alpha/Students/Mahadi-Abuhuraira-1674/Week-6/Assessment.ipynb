### Task 1

Build a classification model that predicts whether a patient has cancer or  not based on the given data.

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

from sklearn.datasets import load_breast_cancer as lbc 
from sklearn.model_selection import train_test_split 
from sklearn.neighbors import KNeighborsClassifier 
```

###### Load the datatset

```python
data = lbc(as_frame = True)
df = data.frame
df.head()
```

###### View the statistical summary for columns

```python
df.describe()
```

```python
df.columns
```

Since about 10 columns are only representing a supposed error occuring in a feature we can drop them to use the columns depicting the properties of the cancer cell with the assumption that only true values will be supplied to the model

```python
error_columns = [col for col in df.columns if 'error' in col] 
error_columns
```

```python
df.drop(columns = error_columns, inplace = True) 
df.head() 
```

###### visualize the 20 columns distribution using a box plot 

```python
%matplotlib inline 
fig = plt.figure()

axs = []
for i in range(4):
    axs.append(fig.add_subplot(2,2,i+1))
    df.iloc[:,i*5:i*5 + 5].plot(kind = 'box',
                               ax = axs[i],
                               figsize = (50,25))
    
plt.show() 
```

From the figure above it's clear that some columns(2 leftmost plots) are distibuted over large values while others(2 rightmost plots) are ranged within a 1 digit value, hence for optimization we will have to normalize the values

###### split the dataset into train_test

```python
x_cancer, y_cancer = df.iloc[:,:-1], df.target 
x_train, x_test, y_train, y_test = train_test_split(x_cancer,y_cancer) 
x_train.shape 
```

###### Normalize the columns using MinMaxScaler object to avoid irregular weight in model

```python
from sklearn.preprocessing import MinMaxScaler 
sc = MinMaxScaler() 
x_train_sc = sc.fit_transform(x_train) 
x_test_sc = sc.fit_transform(x_test) 
```

###### Creating a KNN classifier and fitting the normalized features for training

```python
knn = KNeighborsClassifier(n_neighbors = 5) 
knn.fit(x_train_sc, y_train)
knn.score(x_train_sc,y_train) 
```

```python
 knn.score(x_test_sc,y_test)
```

from the output above it shows that the model has 95% accuracy level in predicting the test data...

count the exact number of columns that were predicted correctly

```python
A = knn.predict(x_test_sc)
comp = pd.DataFrame({'Y':y_test,'Y_pred':A}) 

(comp.iloc[:,0] == comp.iloc[:,1]).value_counts()
```

Using numpy to generate random values and predict using the model

```python
example = np.array([np.random.rand() for x in range(20)]).reshape(1,-1)
print('The cancer is of type: ', data.target_names[knn.predict(example)[0]]) 
```
