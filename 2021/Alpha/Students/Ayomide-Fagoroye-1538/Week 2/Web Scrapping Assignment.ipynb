```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
req = requests.get('http://books.toscrape.com/catalogue/page-1.html')
sp = bs(req.content)

print(sp.prettify())
```

```python
namelist = []
imageurllist = []
pricelist = []
ratinglist = []
for i in range(1,51):
    req = requests.get('http://books.toscrape.com/catalogue/page-'+str(i)+'.html')
    sp = bs(req.content)
    productlist=sp.body.find("section").find_all('article', attrs={'class':'product_pod'})
    
    for product in productlist:
        namelist.append(product.find('img').get('alt'))
        imageurllist.append(product.find('img').get('src'))
        pricelist.append(product.find('p', attrs={'class':'price_color'}).string)
        ratinglist.append(product.select('div ~ p')[0].get('class')[1])
```

```python
prod = {
    'Name': namelist,
    'ImageUrl' : imageurllist,
    'Price' : pricelist,
    'Rating' : ratinglist
}
```

```python
df = pd.DataFrame(prod)
df
```
